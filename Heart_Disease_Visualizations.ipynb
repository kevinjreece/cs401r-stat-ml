```python
ax = data_mat[data_mat.num == 0].plot(kind='scatter', x='thalach', y='chol', color='blue')
data_mat[data_mat.num != 0].plot(kind='scatter', x='thalach', y='chol', color='red', ax=ax)
```

```python
for col1 in data_mat:
    for col2 in data_mat:
        print col1, col2
        ax = data_mat[data_mat.num == 0].plot(kind='scatter', x=col1, y=col2, color='green', s=50)
        data_mat[data_mat.num == 1].plot(kind='scatter', x=col1, y=col2, color='blue', s=50, ax=ax)
        data_mat[data_mat.num == 2].plot(kind='scatter', x=col1, y=col2, color='yellow', s=50, ax=ax)
        data_mat[data_mat.num == 3].plot(kind='scatter', x=col1, y=col2, color='orange', s=50, ax=ax)
        data_mat[data_mat.num == 4].plot(kind='scatter', x=col1, y=col2, color='red', s=50, ax=ax)
        plt.show()
```
