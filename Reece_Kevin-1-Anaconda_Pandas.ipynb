# Part 1: Random Image

```python
import cairo
import numpy
import random

random.seed()
 
# A simple function to display an image in an ipython notebook
def nbimage( data ):
    from IPython.display import display, Image
    from PIL.Image import fromarray
    from StringIO import StringIO
 
    s = StringIO()
    fromarray( data ).save( s, 'png' )
    display( Image( s.getvalue() ) )
 
WIDuTH = 512
HEIGHT = 288
 
# this is a numpy buffer to hold the image data
data = numpy.zeros( (HEIGHT,WIDTH,4), dtype=numpy.uint8 )
 
# this creates a cairo context based on the numpy buffer
ims = cairo.ImageSurface.create_for_data( data, cairo.FORMAT_ARGB32, WIDTH, HEIGHT )
cr = cairo.Context( ims )
 
for x in xrange(0, 500):
    
    r = random.random()
    g = random.random()
    b = random.random()
    r = r - 0.25 if r < 0.5 else r + 0.25
    g = g - 0.25 if g < 0.5 else g + 0.25
    b = b - 0.25 if b < 0.5 else b + 0.25
                
    
    a = random.random()*10
    w = random.random()*10
    x1 = random.random()*WIDTH
    y1 = random.random()*HEIGHT
    x2 = random.random()*WIDTH
    y2 = random.random()*HEIGHT
    x3 = random.random()*WIDTH
    y3 = random.random()*HEIGHT
    x4 = random.random()*WIDTH
    y4 = random.random()*HEIGHT
    
    # draw a random line
    cr.set_source_rgba(r,g,b,a)
    cr.set_line_width(w)
    cr.move_to(x1,y1)
    cr.line_to(x2,y2) if random.random() < 0.5 else cr.curve_to(x2,y2,x3,y3,x4,y4)
    cr.stroke()

# display the image
nbimage( data )
```
